```python
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
data1 = pd.read_csv('../data-set/pregnancy.csv')
data2 = pd.read_csv('../data/patients.csv')
```

```python
types1 = data1[['START', 'STOP', 'Id','DESCRIPTION']]
types2 = data2[['MARITAL', 'RACE', 'ETHNICITY', 'BIRTHDATE', 'DEATHDATE']]
```

# Overall Analysis

##Patient Group

Groups paitients by their IDs to identify those who have had multiple pregnancies/admittance. This could be used later on to track them through each pregnancy to see if their is a decline in pregnancy viability or complications.

```python
patient_group = types1.groupby('Id')
patient_group.size()
```

Extension of previous data set with DESCRIPTION included to better see the multiple entries are because of multiple entries or because of change in status during the admittance period.

```python
category_group = types1.groupby(['Id', 'DESCRIPTION'])
category_group.head()
```

Can be used to see if there is a trend in pregnancy types during years or months.

```python
stats_preg = data1['DESCRIPTION'].value_counts().keys()
nums_preg = data1['DESCRIPTION'].value_counts()
```

```python
type_list = ["Normal pregnancy", "Miscarriage in first trimester", "Non-low risk pregnancy", "Preeclampsia", "Blighted ovum", "Antepartum eclampsia", "Tubal pregnancy", "Fetus with unknown complications"]
data_list = data1[data1.DESCRIPTION.isin(type_list)]
data_list.head()
data_nums = data_list['DESCRIPTION'].value_counts()
```

```python
pregnancy_patterns = data1[['DESCRIPTION', 'START']]
pregnancy_patterns.head()
```

## Pregnancy Distributions by Type

```python
plt.title("Types of Pregnancies")
plt.ylabel("Number of Pregnancies")
plt.xlabel("Pregnancy Descriptions")
plt.bar(stats_preg, nums_preg, color='bgrcmyk')
plt.xticks(rotation=45)
plt.show()
```

```python

fig, ax = plt.subplots()
ax.pie(nums_preg, labels=stats_preg, autopct='%1.1f%%')
ax.axis('equal')
ax.set_title('Types of Pregnancies')
```

## Marital Status of Patients

```python
stats_marr = data2['MARITAL'].value_counts().keys()
nums_marr = data2['MARITAL'].value_counts()
labels_marr = ['Married', 'Single']
```

```python
plt.title("Marital Status of Patients")
plt.ylabel("Number of Patients")
plt.xlabel("Marital Status")
plt.bar(stats_marr, nums_marr, color='bgrcmyk')
plt.show()
```

```python
fig, ax = plt.subplots()
pie = ax.pie(nums_marr, labels = labels_marr, autopct='%1.1f%%')
ax.axis('equal')
ax.set_title('Marital Status of Patients')
```

## Race of Patients

```python
stats_race = data2['RACE'].value_counts().keys()
nums_race = data2['RACE'].value_counts()
```

```python
explode = (0.05, 0.1, 0.1, 0.2, 0.3)
fig, ax = plt.subplots()
ax.pie(nums_race, explode = explode, labels = stats_race, autopct='%1.1f%%')
ax.axis('equal')
ax.legend()
ax.set_title('Race of Patients')
```

```python
plt.title("Race of Patients")
plt.ylabel("Number of Patients")
plt.xlabel("Race")
plt.bar(stats_race, nums_race, color='bgrcmyk')

plt.show()
```

## Ethnicity of Patients

```python
stats_eth = data2['ETHNICITY'].value_counts().keys()
nums_eth = data2['ETHNICITY'].value_counts()
```

```python
fig, ax = plt.subplots()
ax.pie(nums_eth, labels = stats_eth, autopct='%1.1f%%')
ax.axis('equal')
ax.legend()
ax.set_title('Ethnicity of Patients')
```

```python
plt.title("Ethnicity of Patients")
plt.ylabel("Number of Patients")
plt.xlabel("Ethinicity")
plt.bar(stats_eth, nums_eth, color='bgrcmyk')
plt.show()
```

## Diversity of Patients

```python
data2['DIVERSITY'] = data2['RACE'].str.cat(data2['ETHNICITY'], sep= " ")
stats_diverse = data2['DIVERSITY'].value_counts().keys()
nums_diverse = data2['DIVERSITY'].value_counts()
```

```python
fig, ax = plt.subplots()
ax.pie(nums_diverse, labels = stats_diverse, autopct='%1.1f%%')
ax.axis('equal')
ax.legend()
ax.set_title('Diversity of Patients')
```

```python
plt.title("Diversity of Patients")
plt.ylabel("Number of Patients")
plt.xlabel("Diversity")
plt.bar(stats_diverse, nums_diverse, color='bgrcmyk')
plt.xticks(rotation=45)
plt.show()
```
